### 3) 지뢰 찾기 (안전 지대 확인) O은 안전한 곳, X는 지뢰가 매설된 곳, 지뢰의 위치 좌표[행,열]를 출력하세요
* 출력값 :
* 지뢰가 발견되었습니다. 
* 위치 > [0,1], [2,0], [2,2]

````python
#3번
bumb = [
    ['O','X','O'],
    ['O','O','O'],
    ['X','O','X']
  ];
a = ""

for i in range(len(bumb)):
    for j in range(len(bumb[i])):
        if bumb[i][j] == 'X':
            a += f"[{i},{j}]"
print(f"위치 > ({a})")
          
````

### 8) 문자 F의 인덱스값을 찾으시오.
 * 출력값 :
 * 인덱스 값 : (1,2)

````python

# 8번문제
a = [
  ['A', 'B', 'C'],
  ['D', 'E', 'F'],
  ['G', 'H', 'I']
]

for i in range(len(a)):
    for j in range(len(a[i])):
        if a[i][j] == 'F':
          print(f"인덱스 값: ({i},{j})")
          print(f"위치값: ({i+1},{j+1})")
````
